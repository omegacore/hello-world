```python
import os
print os.getcwd()
import lxml.etree as etree
import xtiTranslator as xti

#Change the filename
filename='HXTD.xti'

#Do stuff to clean up the xml tree
tree = etree.parse(filename)
root = tree.getroot()
root = xti.CleanBoxes(root, vlvl=0)
root = xti.SimplifyPdos(root, vlvl=0)

#Print topology if you want
xti.PrintTopology(root)

#Set verbosity (second arg) higher to watch it work
iqList = xti.IQList(root, 0)

for thing in iqList:
    print thing
```
